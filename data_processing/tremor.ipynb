```python
from feature_extraction_RT import FeatureExtractor_RT
import pandas as pd
import os
import numpy as np
```

```python
trainingInfo = "/home/varsallz/Diagnostic_Aid_PD/code/Algorithm/tremor/TrainingData_RestingTremor.csv"
df = pd.read_csv(trainingInfo, delimiter=";")
new_df = df[["SubjectID", "WristAccelerometry", "Disease"]].copy()

validationInfo = "/home/varsallz/Diagnostic_Aid_PD/code/Algorithm/tremor/ValidationData_RestingTremor.csv"
df1 = pd.read_csv(validationInfo, delimiter=";")
new_df1 = df1[["SubjectID","WristAccelerometry", "Disease"]].copy()

# df_total= pd.concat([new_df, new_df1], axis=0)
new_df["label"] = new_df["Disease"].apply(lambda x: 0 if x == "ET" else 1)
new_df1["label"] = new_df1["Disease"].apply(lambda x: 0 if x == "ET" else 1)

array = new_df["WristAccelerometry"].astype(str).values
cleaned_array = [item.replace("./TrainingData/Wrist/", "") for item in array]
new_df["WristAccelerometry"] = cleaned_array

array1 = new_df1["WristAccelerometry"].astype(str).values
cleaned_array1 = [item.replace("./ValidationData/Wrist/", "") for item in array1]
new_df1["WristAccelerometry"] = cleaned_array1

# list2= new_df['WristAccelerometry'].to_list()
# list2
new_df1['WristAccelerometry']
```

```python
dataframes = {}
folder_path = "/home/varsallz/Diagnostic_Aid_PD/data/tremor/UHC/Wrist_Training"

for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path, delimiter=";")
        f =FeatureExtractor_RT(df)
        feature_df = f.process_dataframe(df)

        if file in new_df["WristAccelerometry"].values:
            value = new_df.loc[new_df["WristAccelerometry"] == file, "label"].values[0]
            feature_df["label"] = value
        
        dataframes[file] = feature_df  

final_df = pd.concat(dataframes, ignore_index=True)
final_df
```

```python
dataframes1 = {}
folder_path1 = "/home/varsallz/Diagnostic_Aid_PD/data/tremor/UHC/Wrist_Validation"

for file in os.listdir(folder_path1):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path1, file)
        df1 = pd.read_csv(file_path, delimiter=";")
        f =FeatureExtractor_RT(df1)
        feature_df = f.process_dataframe(df1)

        # # Assign label based on the 'Disease' column
        if file in new_df1["WristAccelerometry"].values:
            value1 = new_df1.loc[new_df1["WristAccelerometry"] == file, "label"].values[0]
            feature_df["label"] = value1
        
        dataframes1[file] = feature_df 

final_df1 = pd.concat(dataframes1, ignore_index=True)
final_df1
```

```python
# # Combine all processed dataframes
# final_df = pd.concat(dataframes, ignore_index=True)
# final_df1 = pd.concat(dataframes1, ignore_index=True)
final_df2 = pd.concat([final_df, final_df1], axis=0, ignore_index=True)

final_df2= final_df2.sample(frac=1, random_state=42).reset_index(drop=True)
final_df2.to_csv("RT_features.csv", index=False)
```
